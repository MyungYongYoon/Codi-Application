```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import PIL
import os

%matplotlib inline

def show_img_compar(img_1, img_2 ):
    f, ax = plt.subplots(1, 2, figsize=(6,6))
    ax[0].imshow(img_1)
    ax[1].imshow(img_2)
    ax[0].axis('off') #hide the axis
    ax[1].axis('off')
    f.tight_layout()
    plt.show()

img = cv.imread("test/(166)IMG_1.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
g = 1.2
out = img.copy()
out = out.astype(np.float)
out = ((out / 255) ** (1 / g)) * 255
out = out.astype(np.uint8)
img = out

h = img.shape[0]
w = img.shape[1]

# img_2 = cv.imread("test/0.492188 0.63241 0.205208 0.385681.jpg")
# img_2 = cv.cvtColor(img_2, cv.COLOR_BGR2RGB)


img1_ann={'top' : [0.553125, 0.687617, 0.23625, 0.210131], 'bottom' : [0.53125, 0.883677, 0.1975, 0.232645]}

x1 = round((img1_ann['top'][0] - (img1_ann['top'][2]/2))*w)
x2 = round((img1_ann['top'][0] + (img1_ann['top'][2]/2))*w)
y1 = round((img1_ann['top'][1] - (img1_ann['top'][3]/2))*h)
y2 = round((img1_ann['top'][1] + (img1_ann['top'][3]/2))*h)
x11 = round((img1_ann['bottom'][0] - (img1_ann['bottom'][2]/2))*w)
x22 = round((img1_ann['bottom'][0] + (img1_ann['bottom'][2]/2))*w)
y11 = round((img1_ann['bottom'][1] - (img1_ann['bottom'][3]/2))*h)
y22 = round((img1_ann['bottom'][1] + (img1_ann['bottom'][3]/2))*h)
print(x1,x2,y1,y2)

top = img[y1+10:y2-10, x1+10:x2-10].copy()
bot = img[y11+10:y22-10, x11+10:x22-10].copy()

# dim = (500, 500)
# # resize image
# img = cv.resize(img, dim, interpolation = cv.INTER_AREA)
# top = cv.resize(top, dim, interpolation = cv.INTER_AREA)
 
show_img_compar(top, bot)
```

```python
from sklearn.cluster import KMeans

clt = KMeans(n_clusters=4)
clt.fit(img.reshape(-1, 3))
    
from collections import Counter
 
def palette_perc(k_cluster, ver=0):
    width = 300
    palette = np.zeros((50, width, 3), np.uint8)
    
    n_pixels = len(k_cluster.labels_)
    counter = Counter(k_cluster.labels_) # count how many pixels per cluster
    perc = {}
    for i in counter:
        perc[i] = np.round(counter[i]/n_pixels, 2)
    perc = dict(sorted(perc.items()))
    
    #for logging purposes
    
    
    step = 0
    
    for idx, centers in enumerate(k_cluster.cluster_centers_): 
        palette[:, step:int(step + perc[idx]*width+1), :] = centers
        step += int(perc[idx]*width+1)
        
    if ver == 0:   
        print(perc)
        print(k_cluster.cluster_centers_)
        return palette
    else: 
        tmp = []
        for i in perc:
             tmp.append(perc[i])
        return k_cluster.cluster_centers_[tmp.index(max(tmp))]
        
clt_1 = clt.fit(top.reshape(-1, 3))

# KMeans(n_clusters=7)

show_img_compar(top, palette_perc(clt_1,0))
print(palette_perc(clt_1,1))
clt_2 = clt.fit(bot.reshape(-1, 3))
show_img_compar(bot, palette_perc(clt_2,0))
```

```python
# 레이블링

root_dir = 'C:/Users/HeoSeokYong/yolo/yolov5/'
os.chdir(root_dir)
ann_path = 'runs/detect/exp3/labels/'
#os.chdir(folder_path)
files = os.listdir(ann_path)
color_dir = root_dir + 'colors_label.txt'
print(len(root_dir.split('/')[-1]))
# color label 받아오기
label = open(color_dir, 'r').read().split('\n')
colors = []
for la in label:
    colors.append(la.split(','))
for c in colors:
    tmp = []
    tmp.append(int(c[2][0:2],16))
    tmp.append(int(c[2][2:4],16))
    tmp.append(int(c[2][4:6],16))
    c[2] = tmp

# for file in files:
#     f = open(root_dir + ann_path + file, 'r')
#     while True:
#         lines = f.readline()
#         if not lines:
#             break
#         print(lines)
img = cv.imread(root_dir + "test/(279)IMG_1.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
g = 1.0
out = img.copy()
out = out.astype(np.float)
out = ((out / 255) ** (1 / g)) * 255
out = out.astype(np.uint8)
img = out

f = open(root_dir + 'runs/detect/exp8/labels/(279)IMG_1.txt', 'r')
tmp=[]
while True:
    lines = f.readline()
    if not lines:
        break
    tmp.append(lines.strip('\n').split(' '))
print(tmp[0][0])
    
    
t = [float(i) for i in tmp[0]][1:5]
b = [float(i) for i in tmp[1]][1:5]


img1_ann={'top' : t, 'bottom' : b}
img = cv.resize(img, (1000,1000), interpolation = cv.INTER_AREA)
x1 = round((img1_ann['top'][0] - (img1_ann['top'][2] * 0.4)) * 1000)
x2 = round((img1_ann['top'][0] + (img1_ann['top'][2] * 0.4)) * 1000)
y1 = round((img1_ann['top'][1] - (img1_ann['top'][3] * 0.4)) * 1000)
y2 = round((img1_ann['top'][1] + (img1_ann['top'][3] * 0.4)) * 1000)
x11 = round((img1_ann['bottom'][0] - (img1_ann['bottom'][2] * 0.4)) * 1000)
x22 = round((img1_ann['bottom'][0] + (img1_ann['bottom'][2] * 0.4)) * 1000)
y11 = round((img1_ann['bottom'][1] - (img1_ann['bottom'][3] * 0.4)) * 1000)
y22 = round((img1_ann['bottom'][1] + (img1_ann['bottom'][3] * 0.4)) * 1000)

top = img[y1:y2, x1:x2].copy()
bot = img[y11:y22, x11:x22].copy()

show_img_compar(top, bot)

def what_color(rgb):
    idx = 0
    lat_sum = 200000
    gray_check = abs(np.mean(rgb) - rgb[0]) + abs(np.mean(rgb) - rgb[1]) + abs(np.mean(rgb) - rgb[2])
    if gray_check < 30 and 720 > sum(rgb) > 110:  # Gray
        return 8
    elif gray_check < 30:  # Black
        if sum(rgb) <= 110:
            return 0
        if sum(rgb) >= 720:  # White
            return 1
    for i in range(len(colors)):
        tsum = 0
        for j in range(3):
            tsum += (colors[i][2][j]-rgb[j])**2
        if tsum < lat_sum:
            if colors[i][0] == '3': # Gray 예외
                continue
            else:
                lat_sum = tsum
                idx = i
    return idx

print(colors[what_color([84.99356369 ,119.40277778, 151.9397019])][1])
```
